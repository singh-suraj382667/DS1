```python
def outliersiqr(df,i):
    if type(df[i][0])!=str:
        q1= df[i].quantile(0.25)
        q3= df[i].quantile(0.75)
        iqr = q3-q1
        low = q1 - (1.5*iqr)
        upp = q3 + (1.5*iqr)
        vl = df[(df[i]>=low) & (df[i]<=upp)]
        df[i] = df[i].apply(lambda x : df[i].mean() if x<= low else df[i].mean() if x>=upp else x )
```

```python
def outliersmst(df,i):
    if type(df[i][0])!=str:
        m= df[i].mean()
        sd = df[i].std()
        low = m - (3*sd)
        upp = m + (3*sd)
        ovalid = df[df[i]>=low & (df[i]<= upp)]
        df[i] = df[i].apply(lambda x : df[i].mean() if x<= low else df[i].mean() if x>=upp else x )
```

```python
def encode(df,i):
    if type(df[i][0])==str:
        d =one.fit_transform(df[[i]]).toarray()
        cname = one.categories_
        dd = pd.DataFrame(d, columns=cname[0][1:: ], dtype = int)
        fdf= pd.concat([df,dd], axis = 1)
        ffdf = fdf.drop([i], axis = 1)
        return ffdf
```
